# ABSTRACTION


```python
class coffeeMachine:
  def __init__(self):
    self.water = 100

  def make_coffee(self):
    if self.water >= 10:
        self.water -= 10
      print("coffee is ready!")
    else:
      print("not enough water to make coffee.")

machine =  coffeeMachine()

machine.make_coffee()
```

# DATA ENCAPSULATION

```python
class BankAccount:
  def __init__(self):
    self.__balance = 0

  def set_balance(self, amount):
     self.__balance = amount

  def get_balance(self):
     return self.__balance

  def withdraw(self, amount):
    if self.__balance >= amount:
      self.__balance -= amount
      print(f"Withdrawal of {amount} successful. Remaining balance: {self.__balance}")


account = BankAccount()
account.set_balance(1000)
print("Account balance:" , account.get_balance())

account.withdraw(400)
```

# INHERITANCE
 TWO TYPES : SINLGE
              MULTIPLE

```python
class Animal:
  def move(self):
    print("Animal is moving")

class Dog(Animal):
  def bark(self):
    print("Dog is barking")

class Cat(Animal):
  def meow(self):
    print("Cat is meowing")

dog = Dog()
dog.move()
dog.bark()

cat = Cat()
cat.move()
cat.meow()
```

# SINGLE LEVEL

```python
class Vehicle:
    def move(self):
        print("Vehicle is moving")

class Car(Vehicle):
    def honk(self):
        print("Car is honking")



car = Car()
car.move()
car.honk()

```

# POLYMORPHISM

```python
class cat:
  def sound(self):
    print("meow")

class dog:
  def sound(self):
    print("bark")

for animal in (cat(), dog()):
  animal.sound()
```
