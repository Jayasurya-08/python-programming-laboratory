```python
import nbformat
import os
from pathlib import Path

# --- 1) Exact filenames to merge (as shown in your GitHub image) ---
notebooks_order = [
    "GROCERY_SYSTEM.ipynb",
    "MED23CI202_(1).ipynb",
    "PLOTS AND OPPS.ipynb",
    "crud.ipynb",
    "data visualization.ipynb",
    "file_operation.ipynb",
    "fileoperation II.ipynb",
    "opps 2.ipynb"
]

# --- 2) Verify which files exist ---
existing = []
missing = []
for p in notebooks_order:
    if Path(p).exists():
        existing.append(p)
    else:
        missing.append(p)

print("âœ… Found notebooks:")
for p in existing:
    print(" -", p)
if missing:
    print("\nâš  Missing (check spelling/case):")
    for p in missing:
        print(" -", p)

# --- 3) Merge them into one file ---
merged = nbformat.v4.new_notebook()
merged.cells = []

for nb_file in existing:
    merged.cells.append(nbformat.v4.new_markdown_cell(f"## ðŸ“˜ Content from {nb_file}"))
    with open(nb_file, "r", encoding="utf-8") as f:
        nb = nbformat.read(f, as_version=4)
        merged.cells.extend(nb.cells)

# --- 4) Save the merged notebook ---
output_file = "Combined_Notebook.ipynb"
with open(output_file, "w", encoding="utf-8") as f:
    nbformat.write(merged, f)

print(f"\nâœ… Successfully created: {output_file}")
```
