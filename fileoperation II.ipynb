# TELL OPERATION

```python
with open("/content/hello world! python programming lab.txt", "r") as f:
  print(f.tell())
  data = f.read(6)
  print(data)
  print(f.tell())
```

# SEEK

```python
with open("/content/hello world! python programming lab.txt", "r") as f:
  print(f.seek(0))
  data = f.read(4)
  print(data)
  print(f.seek(0))
```

# TRUE & FALSE

```python
with  open("/content/hello world! python programming lab.txt", "r") as f:
  print(f.readable())
  print(f.writable())
  print(f.seekable())
```
