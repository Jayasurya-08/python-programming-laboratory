# FILE HANDLING
READ FILE

```python
file_path = r"/content/hello world! python programming lab.txt"
print("=== example 1: Read Entire File ===")
with open(file_path, "r") as file:
    content = file.read()
    print(content)
```

 # read line by line

```python
file_path = r"/content/hello world! python programming lab.txt"
print("=== Hello World!: Read Line by Line ===")
with open(file_path, "r") as file:
    for line in file:
        print("Line from file: ", line.strip())
```

```python
file_path = r"/content/hello world! python programming lab.txt"
print(" === Read specific ===")
with open(file_path, "r") as file:
    first_14_chars = file.read(14)
    print("First 14 characters:", first_14_chars)

    file.seek(0)
    file.readline()
    first_line = file.readline()
    print("first line:", first_line.strip())
```

# write file

```python
file_path = r"/content/hello world! python programming lab.txt"

with open(file_path, "a") as file:
  file.write(" this line is append later.\n")
print("new line append successfully")
```

# ADDING MULTIPLE LINE

```python
file_path = r"/content/hello world! python programming lab.txt"

lines = [
    "line 1 = New line\n",
    "line 2 = New line written\n",
    "line 3 = New line written successfully\n"
]

with open(file_path, "a") as file:
  file.writelines(lines)
print("multiple lines appended successfully")
```
