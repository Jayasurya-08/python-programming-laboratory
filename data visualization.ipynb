# 1. BAR PLOT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "student": ["Ruba","Surya","Tanu","Laksha"],
    "fee paid": [250000, 175000, 70000, 43000]
}

df=pd.DataFrame(data)
colors=["red","blue","black","green","yellow"]
sns.barplot(x="student", y="fee paid", data=df, palette=colors)
plt.title("Student Fee Data")
plt.show()
```

# TYPES OF DISTRIBUTION PLOT
* HISTOGRAM
* BOX PLOT
* DENSITY PLOT

# * BOX PLOT

```python
fees = [250000, 175000, 70000, 43000]
plt.boxplot(fees)
plt.show()
```

# 3.DENSITY PLOT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "student": ["Ruba","Surya","Tanu","Laksha"],
    "fee paid": [250000, 175000, 70000, 43000]
}
density=sns.kdeplot(data["fee paid"],shade=True)

df=pd.DataFrame(data)
colors=["red","blue","black","green","yellow"]

plt.title("Student Fee Data")
plt.show()
```
