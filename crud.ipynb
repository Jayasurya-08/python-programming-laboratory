NESTED DICTIONARY

```python
student= {
    "Ruba" :{ "math" : 85, "science" : 90, "english" : 100},
    "Surya" :{ "math" : 89, "science" : 95, "english" : 95},
    "Tanu" :{ "math" : 80, "science" : 92, "english" : 80},
    "Sriram" :{ "math" : 100, "science" : 87, "english" : 98},
    "Ganesh" :{ "math" : 81, "science" : 96, "english" : 90},
}
print(student)
```

for data manipulation using dictionary convert the dictionary to dataframe

```python
import pandas as pd
data = pd.DataFrame(student)
data
```

```python
data = pd.DataFrame(student).transpose()
data
```

Add analysis to the colunms

```python
data["average"] = data.mean(axis=1) # axis = 1 denotes row wise access to the dataframe
data
```

```python
data["math"] = data.mean(axis=0) # axis = 0denotes row wise access to the dataframe
data
# the result is not a number (Nan) becoz there is no variable to store the average of
```

DATA VISUALIZATIONS

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
data["average"].plot(kind="bar", color= "ORANGE", edgecolor= "black")
plt.title("Average marks of students")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()
```

HOW TO VISULAISE SUBJECTWISE TREND

```python
plt.figure(figsize=(8,5))
#data["average"].plot(kind="bar", color= "ORANGE", edgecolor= "black")
for subject in ["math" , "science" , "english"]:
    plt.plot(data[subject], marker="*", label=subject)
plt.title("subject wise performance")
plt.xlabel("Students")
plt.ylabel(" Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
data[["math", "science", "english"]].plot(kind="box")
plt.title("subject wise performance")
plt.xlabel("Students")
plt.ylabel(" Marks")
plt.show()
```

2. The nested dictionary is coverted to pandas dataframe
3. The data analytics are performed and visulised using matplotlib.
